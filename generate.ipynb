```python
new_dataset = []
for dat in (data['train']):
    # print(dat)
    new = deepcopy(dat)
    if '####' in new['rejected']:
        pre = new['rejected'][:new['rejected'].index('####')]
        pre_ans = new['rejected'][new['rejected'].index('####') + 4:]
        pre_ans = pre_ans.strip()
        new['rejected'] = pre + 'The answer is: ' + pre_ans
        new_dataset.append(new)
    elif r'\boxed{' in  new['rejected'] and '}' in new['rejected'] and '<|endoftext|>' in new['rejected']:
        # print(1)
        pre = new['rejected'][new['rejected'].rindex(r'\boxed{') + 7:new['rejected'].rindex('}')]
        
        new['rejected'] = new['rejected'][:new['rejected'].index('<|endoftext|>')] + '\nThe answer is: ' + pre + '<|endoftext|>'
        print(1)
        new_dataset.append(new)
    else:
        print('skip')
        

    
print(new_dataset[0]['rejected'])
```
